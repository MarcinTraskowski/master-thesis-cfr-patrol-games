```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sns

with open("../results/local/demo_1/strategy_best_5_initializations.pkl", "rb") as f:
    P = pickle.load(f)

P = np.array(P)

plt.figure(figsize=(5.4, 5.4))

ax = sns.heatmap(
    P,
    cmap="rocket_r",
    square=True,
    vmin=0,
    vmax=1,
    linewidths=0.2,
    linecolor="white",
    cbar_kws={
        "ticks": [0,0.2,0.4,0.6,0.8,1],
        "fraction": 0.045,
        "pad": 0.015,
    }
)

ax.xaxis.tick_top()
ax.tick_params(top=True, bottom=False, labeltop=True, labelbottom=False)

plt.tight_layout(pad=0.1)
```
